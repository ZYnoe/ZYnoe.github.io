```python
import pandas as pd
import requests
import time
from tqdm import tqdm

# ================= API config =================
API_KEY = ""  # Replace with your Elsevier API key
# ===========================================

def get_elsevier_abstract_v2(doi, api_key):
    
    base_url = f"https://api.elsevier.com/content/abstract/doi/{doi}"
    
    headers = {
        "X-ELS-APIKey": api_key,
        "Accept": "application/json"
    }
    
    try:
        
        response = requests.get(base_url, headers=headers, params={'view': 'FULL'}, timeout=10)
        
        if response.status_code == 200:
            data = response.json()
            
           
            root = data.get('abstracts-retrieval-response', {})
            
            
            abstract = root.get('coredata', {}).get('dc:description')
            
           
            
            if abstract:
                return abstract.strip()
            else:
                return "No Abstract in Record"
                
        elif response.status_code == 400:
            return "Error: 400 (Bad Request - Check API Key Type)"
        elif response.status_code == 401:
            return "Error: 401 (Unauthorized)"
        elif response.status_code == 404:
            return "Error: 404 (DOI Not Found)"
        else:
            return f"Error: {response.status_code}"
            
    except Exception as e:
        return f"Exception: {str(e)}"

def main():
    
    csv_url = "https://raw.githubusercontent.com/M3RG-IITD/MatSciBERT/refs/heads/main/pretraining/piis_dois.csv"
    try:
        df = pd.read_csv(csv_url)
        print(f"Loaded {len(df)} DOIs.")
    except Exception as e:
        print(f"Failed to load CSV: {e}")
        return

    
    
    
    print("Starting download using Abstract API...")
    tqdm.pandas()
    df['abstract'] = df['DOI'].progress_apply(lambda x: get_elsevier_abstract_v2(x, API_KEY))
    
    print("\nResults:")
    print(df[['DOI', 'abstract']])

    
    df.to_csv("abstracts_fixed.csv", index=False)

if __name__ == "__main__":
    main()
```
