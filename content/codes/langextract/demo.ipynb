```python
import json

with open('abstracts.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

# 统计
valid = sum(1 for item in data if item.get('abstract'))
print(f"\n✅ Successful {valid} / In total: {len(data)} / ❌ Failed: {len(data) - valid}\n")
```

```python
import json

with open('abstracts.json', 'r', encoding='utf-8') as f:
    data = json.load(f)
    
    # 只提取有 abstract 的条目
    abstracts = [
        item['abstract'].strip() 
        for item in data 
        if item.get('abstract')
    ]
    
print(abstracts[:5])
print(type(abstracts))
```

```python
for abstract in abstracts[:5]:
    abstract = abstract.replace('\n', ' ')
    print(abstract)
```
